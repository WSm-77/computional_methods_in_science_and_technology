```python
# Create data for 3 lines
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

fig, ax = plt.subplots()
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.5, 1.5)

# Initialize three empty lines
lines = [ax.plot([], [], lw=2)[0] for _ in range(3)]
styles = ['r-', 'g--', 'b:']

def init():
    for line in lines:
        line.set_data([], [])
    return lines  # Return all lines as a tuple

def animate(i):
    # Update data for each line up to frame i
    lines[0].set_data(x[:i], y1[:i])  # Red solid line
    lines[1].set_data(x[:i], y2[:i])  # Green dashed line
    lines[2].set_data(x[:i], y3[:i])  # Blue dotted line

    # Apply line styles (optional)
    for line, style in zip(lines, styles):
        line.set_linestyle(style)

    return lines  # Return all lines as a tuple

ani = FuncAnimation(fig, animate, init_func=init,
                   frames=len(x), interval=20, blit=True)
plt.show()
```
